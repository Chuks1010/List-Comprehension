## Time series data analysis

The data show the unemployment rate in the US and made available by the United States Bureau of Labor Statistics (BLS). 
The BLS surveys, calculates the monthly unemployment rate and  releases the data for download. The unemployment rate is the percentage of individuals in the labor force without a job.

The first thing is to import pandas and explore the data

```python
import pandas as pd
unrate = pd.read_csv('unrate.csv')
```

```python
unrate.info()
```

From the information above pandas set the data type of the date column as a text column and gives it a data type of object indicating a string. This needs to be formatted to reflex the datetime data type by using the pandas.to_datetime function.

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
unrate.info()
```

```python
#looking at the first twelve rows or records
unrate.head(12)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(unrate['DATE'], unrate['VALUE'])
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948 - 2016')
plt.show()
```

```python
unrate['VALUE'].max()
```

```python
fig = plt.figure()
# 2 row by 2 column layout
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['VALUE'])
ax1.set_title('Monthly Unemployment Rate, 1948')
ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['VALUE'])
ax2.set_title('Monthly Unemployment Rate, 1949')
plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
fig = plt.figure(figsize=(6,3))

plt.plot(unrate['MONTH'][0:12], unrate['VALUE'][0:12], c='red')
plt.plot(unrate['MONTH'][12:24], unrate['VALUE'][12:24], c='blue')

plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
fig = plt.figure(figsize=(6,3))

plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='red')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='blue')

plt.show()
```

Visualize 5 years (1948 - 1956) worth of unemployment rates on the same subplot

```python
plt.show()
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
plt.legend(loc='upper left')
plt.xlabel('Month, Integer')
plt.ylabel('Unemployment Rate, Percent')
plt.title('Monthly Unemployment Trends, 1948-1952')

plt.show()
```
